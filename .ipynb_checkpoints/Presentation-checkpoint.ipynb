```julia
using RHEOS
using PyPlot
```

```julia
datat = timeline(t_start=0, t_end=20.0, step = 0.02)
# data = strainfunction(datat,hstep(offset=5.0,amp=5));
data = strainfunction(datat,ramp(offset=2.0,gradient=0.8));
ylabel("Strain")
xlabel("Time")
plot(data.t, data.ϵ, "-", linewidth=4)
```

```julia
maxwell = RheoModel(Maxwell, k = 1, η = 1)
datap = modelpredict(data, maxwell)
```

```julia
plot(datap.t, datap.σ,linewidth=4)
# xlabel("Time")
# ylabel("Stress")
```

```julia
dashpot = RheoModel(Dashpot, η = 1)
datap = modelpredict(data, dashpot)
```

```julia
plot(datap.t, datap.σ,linewidth=4)
```

```julia
kv = RheoModel(KelvinVoigt, η = 1,k = 1)
datap = modelpredict(data, kv)
```

```julia
plot(datap.t, datap.σ,linewidth=4)
```
