```julia
using RHEOS
using PyPlot
```

```julia
datat = timeline(t_start=0, t_end=20.0, step = 0.02)
data = strainfunction(datat,hstep(offset=5.0,amp=10));
# data = strainfunction(datat,ramp(offset=2.0,gradient=0.8));
ylabel("Strain")
xlabel("Time")
plot(data.t, data.ϵ, "-", linewidth=3)
savefig("images/step_strain_amp_10.eps")
```

```julia
kv = RheoModel(KelvinVoigt, k = 10, η = 1)
datap = modelpredict(data, kv)
```

```julia
plot(datap.t, datap.σ,linewidth=3)
xlabel("Time")
ylabel("Stress")
savefig("images/stress_response_step_10.eps")
```
